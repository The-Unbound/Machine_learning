```python
def covert_dur(dur):
    at = str(dur).split(".")
    hours = at[0]
    minutes = at[1]
    return int(hours)*60 + int(minutes)
    
    
```

```python
d["Duration_min"] = d.duration.apply(covert_dur)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
d.stops.unique()
key  = {'zero' : 0, 'one' : 1, 'two_or_more' : 2}
```

```python
d.stops = d.stops.apply(lambda x : key.get(x,0))
```

```python
d.head()
```

```python
target = ["price"]
features = ["airline","flight","source_city","stops","arrival_time","destination_city","class","days_left"]
```

```python
D_lr = ["Duration_min","days_left","stops"]
```

```python
x_lr = d[D_lr]
y_lr= d[target]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_lr,y_lr,test_size = 0.2,random_state= 42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred_lr = lr.predict(x_test)
```

```python
#accuracy
print(f"r2 score of the model is {r2_score(y_test,pred_lr)}")
```

```python
print(f"mean_squared_error : {mean_squared_error(y_test,pred_lr)}")
```
